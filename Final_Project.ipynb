# Part 2:  Ratio Analysis

##Calculating ROA, Current Ratio, Inventory Turnover and Leverage

```python
f_roll = lambda x: x.rolling(2).mean()
df['avg_asset']=df.groupby('tic').at.apply(f_roll)
df['avg_inv']=df.groupby('tic').invt.apply(f_roll)
df['avg_ceq']=df.groupby('tic').ceq.apply(f_roll)
```

```python
df['roa'] = ((df.ni-df.dvp)/df.avg_asset)*100
df['current_ratio'] = (df.act/df.lct)
df['inv_t'] = (df.cogs/df.avg_inv)
df['lvg'] = (df.avg_asset/df.avg_ceq)
```

```python
import numpy as np
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
df= df.dropna(subset = ['roa','current_ratio','inv_t','lvg'])
df
```

# Part 4: Regression Analysis

##Sorting Data

```python
df_stock['fyear'] = df_stock['date'].dt.year
df_stock = df_stock.rename(columns = {'TICKER':'tic'})
reg_data = df.merge(df_stock, on =['tic','fyear'] , how='inner')
reg_data.head()
```

```python
import numpy as np
```

```python
reg_data = reg_data.dropna(subset = ['at','roa','lvg'])
reg_data.head()
```

```python
reg_data['positive'] = np.where(reg_data['stock_ret']>0,1,0)
reg_data.head()
```

```python
reg_data['log_at'] = np.log(reg_data['at']).round(2)
reg_data.head()
```

```python
from scipy.stats.mstats import winsorize
```

```python
reg_data['win_log_at'] = winsorize(reg_data['log_at'], limits=[0.01, 0.01])
reg_data['win_lvg'] = winsorize(reg_data['lvg'], limits=[0.01, 0.01])
reg_data['win_roa'] = winsorize(reg_data['roa'], limits=[0.01, 0.01])
reg_data['win_stock_ret'] = winsorize(reg_data['stock_ret'], limits=[0.01, 0.01])
reg_data.head()
```

```python
reg_data[['log_at','win_log_at']].describe()
```

```python
reg_data[['lvg','win_lvg']].describe()
```

##Regression Table

```python
reg_data[['roa','win_roa']].describe()
```

```python
import statsmodels.api as sm
```

```python
X = reg_data[['win_log_at','win_roa','win_lvg']]
Y = reg_data['win_stock_ret']
```

```python
X
```

```python
X = sm.add_constant(X) 
model = sm.OLS(Y, X).fit()
predictions = model.predict(X)
#add constant 

print_model = model.summary()
print(print_model);
```

The only statistically significant variable in this regression analysis table is ROA because its t-value is 3.646, and P>|t| is smaller than 0.005. The coefficient of win_roa is 0.0177, showing as the ROA increases by 1% the stock return increases by 1.77%. In general, we assume that companies have higher ROA are performing quite well so they will consequently have higher stock return. The positive coefficient of ROA is same as wee have expected. However, R-sqaured value is 0.064, which is considered as a very low value and showing low correlation between ROA and stock return.

# Part 6:  Unsupervised Machine Learning: K-means clustering

```python
# Choose two variables
# ROA and at (total asset)
# We do analysis for only year 2021
```

```python
# Keep only the year 2021
dfcluster = reg_data[reg_data['fyear'] == 2021]
dfcluster.head()
```

```python
# read the csv and keep only tic, at, and roa
dfcluster = reg_data[['tic','win_roa','win_log_at']]
dfcluster
```

```python
dfcluster = dfcluster[dfcluster['win_log_at'] > 0]
dfcluster = dfcluster[dfcluster['win_roa'] > 0]
```

```python
#drop na
dfcluster.dropna()
dfcluster.head()
```

```python
# Describe
dfcluster.describe()
```

```python
#Define the array X
X = dfcluster.iloc[:,1:3].values
X
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

## Elbow method to identify k

```python
#elbow method
wcss = []
for i in range(1,11):
    k_means = KMeans(n_clusters=i,init='k-means++',max_iter=300,  n_init=10,random_state=0)
    k_means.fit(X)
    #kmeans algorithm fits to the X dataset
    wcss.append(k_means.inertia_)
#kmeans inertia_ attribute is:  Sum of squared distances of samples #to their closest cluster center.
#Plot the elbow graph
plt.plot(np.arange(1,11),wcss)
plt.xlabel('Clusters')
plt.ylabel('SSE')
plt.show()
```

## K means clustering

```python
# The optimal number of k is 3
k_means_optimum = KMeans(n_clusters = 3, init = 'k-means++',  random_state=0)
y = k_means_optimum.fit_predict(X)
y
```

```python
dfcluster['cluster'] = y 
dfcluster
```
